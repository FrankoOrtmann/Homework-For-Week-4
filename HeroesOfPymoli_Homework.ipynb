Heroes Analysis

* The average purchase per player is $4.13, males are the only demographic to spend less than that on average.
* All 5 top spenders in the game are Male
* 4 of the 5 top spenders are in the 20-24 age range, the 5th is 25 years old.
* 1 of the 5 top spenders is a Female, the rest are male.

```python
import pandas as pd
import numpy as np

file = "Resources/purchase_data.csv"
pd.options.mode.chained_assignment = None

PurchaseData = pd.read_csv(file)
PD2 = pd.read_csv(file)
```

```python
PurchaseData.columns
```

```python
TotalPlayers = PurchaseData['SN'].nunique()
TPPD = pd.DataFrame({'Total Players': [TotalPlayers]})
TPPD
```

```python
UniqueItems = PurchaseData['Item Name'].nunique()
AveragePrice = PurchaseData['Price'].mean()
TotalPurchases = PurchaseData['Purchase ID'].count()
TotalRev = PurchaseData['Price'].sum()


SummaryDF = pd.DataFrame({'Number of Unique Items': [UniqueItems], 'Average Price': [AveragePrice], 
                          'Number of Purchases': [TotalPurchases], 'Total Revenue': [TotalRev]})

SummaryDF["Average Price"] = SummaryDF["Average Price"].map("${:.2f}".format)
SummaryDF["Total Revenue"] = SummaryDF["Total Revenue"].map("${:,.2f}".format)
SummaryDF
```

```python
GenderGroup = PurchaseData.drop_duplicates(subset=['SN'], inplace=True)
GenderGroup = PurchaseData["Gender"].value_counts()
GenderPD = pd.DataFrame(GenderGroup)
GenderPD = GenderPD.rename(columns={"Gender": "Total Count"})
GenderPDPercent = GenderPD["Total Count"]/GenderPD["Total Count"].sum()*100
GenderPD["Percentage of Players"] = GenderPDPercent
GenderPD["Percentage of Players"] = GenderPD["Percentage of Players"].map("{:.2f}%".format)
GenderPD
```

```python
g = PD2.groupby('Gender').Price.agg(['mean', 'sum', 'count'])
GenderNumbers = pd.DataFrame(g)
GenderSummary = pd.concat([GenderNumbers, GenderPD], sort=True, axis=1)
AvgTtl = GenderSummary['sum']/GenderSummary['Total Count']
GenderSummary["Average Total Purchase per Person"] = AvgTtl
OrganizedGenderSummary = GenderSummary[["count", "mean", "sum", "Average Total Purchase per Person"]]
CleanGS = OrganizedGenderSummary.rename(columns={"count":"Purchase Count","mean":"Average Purchase Price","sum":"Total Purchase Value"})
CleanGS["Average Purchase Price"] = CleanGS["Average Purchase Price"].map("${:.2f}".format)
CleanGS["Total Purchase Value"] = CleanGS["Total Purchase Value"].map("${:,.2f}".format)
CleanGS["Average Total Purchase per Person"] = CleanGS["Average Total Purchase per Person"].map("${:,.2f}".format)
CleanGS
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 500]
AgeGroups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
PurchaseData.drop_duplicates(subset=['SN'], inplace=True)
PurchaseData["Age Range"] = pd.cut(PurchaseData["Age"], bins, labels=AgeGroups)
b = PurchaseData.groupby("Age Range").count()
BinGroup = pd.DataFrame(b)
BinGroup = BinGroup.rename(columns={"Age": "Total Count"})
BinGroup = BinGroup[["Total Count"]]
AgePercent = BinGroup["Total Count"]/BinGroup["Total Count"].sum()*100
BinGroup["Percentage of Players"] = AgePercent
BinGroup["Percentage of Players"] = BinGroup["Percentage of Players"].map("{:.2f}%".format)
BinGroup
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 500]
AgeGroups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
PD2["Age Range"] = pd.cut(PD2["Age"], bins, labels=AgeGroups)
b = PD2.groupby("Age Range").Price.agg(['mean', 'sum', 'count'])
AgeGroup = pd.DataFrame(b)
AgeSummary = pd.concat([AgeGroup, BinGroup], sort=True, axis=1)
AgeSumRenamed = AgeSummary.rename(columns={"mean":"Average Purchase Price","sum":"Total Purchase Value","count":"Purchase Count"})
AgeSummaryClean = AgeSumRenamed[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
AgeAvgTtl = AgeSumRenamed['Total Purchase Value'] / BinGroup['Total Count']
AgeSummaryClean["Avg Total Purchase per Person"] = AgeAvgTtl
AgeSummaryClean["Average Purchase Price"] = AgeSummaryClean["Average Purchase Price"].map("${:.2f}".format)
AgeSummaryClean["Total Purchase Value"] = AgeSummaryClean["Total Purchase Value"].map("${:,.2f}".format)
AgeSummaryClean["Avg Total Purchase per Person"] = AgeSummaryClean["Avg Total Purchase per Person"].map("${:,.2f}".format)
AgeSummaryClean
```

```python
TSGroup = PD2.groupby("SN").Price.agg(['count', 'mean', 'sum', ])
TopSpenders = pd.DataFrame(TSGroup)
TSRenamed = TopSpenders.rename(columns={"mean":"Average Purchase Price","sum":"Total Purchase Value","count":"Purchase Count"})
TSRenamed = TSRenamed.sort_values("Total Purchase Value", ascending=False)
TSRenamed["Average Purchase Price"] = TSRenamed["Average Purchase Price"].map("${:.2f}".format)
TSRenamed["Total Purchase Value"] = TSRenamed["Total Purchase Value"].map("${:,.2f}".format)
TSRenamed.head()
```

```python
p = PD2.groupby(["Item ID", "Item Name"]).Price.agg(['count', 'mean', 'sum', ])
Pop = pd.DataFrame(p)
PopRenamed = Pop.rename(columns={"mean":"Average Purchase Price","sum":"Total Purchase Value","count":"Purchase Count"})
PopRenamed = PopRenamed.sort_values("Purchase Count", ascending=False)
PopRenamed["Average Purchase Price"] = PopRenamed["Average Purchase Price"].map("${:.2f}".format)
PopRenamed["Total Purchase Value"] = PopRenamed["Total Purchase Value"].map("${:,.2f}".format)
PopRenamed.head()
```
